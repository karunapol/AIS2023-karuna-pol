```python
import pandas as pd
df = pd.read_csv("C:/Users/Mac/Indprj/task1.csv") 
df
```

```python
task= {"jan": 800, "feb": 500, "march": 1000,"apr":1200,"may":1500,"jun":900,"jul":200,"aug":1300,"sep":700,"oct":2000,"nov":1800,"dec":600}
for a in task:
    
    print(a,'=',task[a])
#task
```

```python
def t1 (capacity,demand):
    cap=capacity
    dem=demand
    demo=df['Cap'].loc[df.index[1]]
    df['Cap']=df['Cap'].replace({demo:cap})
    
    key1=dem.keys()
    value1=list(dem.values())
    
    for i in range(len(key1)):
        df.loc[i,'Demand']=value1[i]
        sub=cap-value1[i]
        if sub<0:
            df.loc[i,'ExtraQty']=abs(sub)
            df.loc[i,'Remainingcap']=0

        elif sub>0:
            df.loc[i,'ExtraQty']=0
            df.loc[i,'Remainingcap']=sub

        else:
            df.loc[i,'ExtraQty']=0
            df.loc[i,'Remainingcap']=0

    df.to_csv("C:/Users/Mac/Indprj/task1.csv", index=False)
    return df
```

```python
t1(100,task)
```
